## **Cyber Threat Intelligence Assignment – Milestone Test**

```python
import json
import requests
```

```python
url = "https://raw.githubusercontent.com/mitre/cti/master/enterprise-attack/enterprise-attack.json"
response = requests.get(url)
data = response.json()
print("Dataset loaded successfully")
```

```python
print("Top-level keys in dataset:")
print(data.keys())
```

```python
techniques = []

for obj in data["objects"]:
    if obj.get("type") == "attack-pattern":
        techniques.append(obj)

print("Total attack techniques extracted:", len(techniques))
```

```python
def threat_score(tech):
    score = 5  # Base score
    name = tech.get("name", "").lower()
    description = tech.get("description", "").lower()

    if "credential" in name:
        score += 3
    if "execution" in name:
        score += 2
    if "privilege" in name:
        score += 3
    if "persistence" in name:
        score += 2
    if "lateral" in name:
        score += 2

    if "administrator" in description:
        score += 2
    if "remote" in description:
        score += 2
    if "bypass" in description:
        score += 2
    if "stealth" in description:
        score += 1

    return score
```

```python
scored_techniques = []

for tech in techniques:
    tech_name = tech.get("name", "Unknown Technique")
    score = threat_score(tech)
    scored_techniques.append((tech_name, score))
```

```python
scored_techniques.sort(key=lambda x: x[1], reverse=True)
print("Techniques sorted by risk score")
print(scored_techniques)
```

```python
print("Top 10 Highest-Risk Techniques:\n")
for tech, score in scored_techniques[:10]:
    print(f"{tech} — Score: {score}")
```

```python
print("\nCritical Threats (Score ≥ 8.9):\n")
for tech, score in scored_techniques:
    if score >= 8.9:
        print(f"{tech} — Score: {score}")
```

### Explanation

This scoring system helps cybersecurity analysts prioritize threats by assigning risk scores based on critical keywords found in both technique names and descriptions. Techniques involving credentials, privilege escalation, remote access, or stealth behavior receive higher scores, indicating higher risk. This allows analysts to focus on the most dangerous attack techniques first, improving investigation efficiency.
